### Idea 1 :
Creating function that will categorize numerical and categorical variable from the input dataset and plot graph for all variables present in that dataset

Input from user - path i.e directory, File name

```python
def graph():
    
    import os
    import pandas as pd
    import numpy as np
    import matplotlib.pyplot as plt
    import seaborn as sns
    
    # Allow user to enter directory to fetch the file
    dir = input("Enter the path  : ")     # e.g : C:\\Users\\lenovo\\Desktop\\Praxis\\Practice_doc (without " ")
    os.chdir(dir)
    
    # Allow user to enter the name of the file
    # user can call the function multiple times for different data
    file = input('Enter the file name')   # e.g cars.csv (without " ")
    df = pd.read_csv(file) 
        
    for i in df:
        if (np.dtype(df[i]) != 'object' and len(np.unique(df[i]>50))):  # Check whether the variable is categorical or numerical 
            # if the datatype of variable is not object then consider it as the numerical variable
            # Histogram
            df.hist(column=i,grid=False,figsize=(10,8))
            plt.savefig(str(i)+'_histogram.png')   # to save the graph as image  
            plt.close()

            # Boxplot
            df.boxplot(column=i,grid=False,notch=True,figsize=(10,8))
            plt.savefig(str(i)+'_boxplot.png')
            plt.close()

        else:
            # Bar graph
            sns.countplot(df[i])
            plt.savefig(str(i)+'_bar_graph.png')
            plt.close()
    
    # Print the statistics i.e means,min,max of the data
    Describe = df.describe()   
    Describe.to_csv("Describe.csv")
    print("Please check the folder")
```

```python
graph()
```

### Idea 2 : Improvisation 1
We will allow user to enter specific columns .
This will help us to observe the plot for only required variables 
If user like to see the graph for all variables present in the dataset, user can type "All" in the input

Input from user : directory, file name, required column names

```python
def graph():
    
    import os
    import pandas as pd
    import numpy as np
    import matplotlib.pyplot as plt
    import seaborn as sns
    
    dir = input("Enter the path  : ")     # e.g : C:\\Users\\lenovo\\Desktop\\Praxis\\Practice_doc (without " ")
    os.chdir(dir)
    
    file = input('Enter the file name')   # e.g cars.csv (without " ")
    df = pd.read_csv(file) 
    
    # Allow user to enter specific column name 
    print("Print All if Column not specified ")
    col = input('Enter the name of column separated by comma : ')
    col1 = list(col.split(","))
    if col1 == ["All"]:
        col = []
        for i in df:
            col.append(i)
    else:
        col = col1
        
    for i in col:
        if (np.dtype(df[i]) != 'object' and len(np.unique(df[i]>50))):
            # Histogram
            df.hist(column=i,grid=False,figsize=(10,8))
            plt.show()
            plt.savefig(str(i)+'_histogram.png')
            plt.close()

            # Boxplot
            df.boxplot(column=i,grid=False,notch=True,figsize=(10,8))
            plt.show()
            plt.savefig(str(i)+'_boxplot.png')
            plt.close()

        else:
            # Bar graph
            sns.countplot(df[i])
            plt.show()
            plt.savefig(str(i)+'_bar_graph.png')
            plt.close()
            
    Describe = df.describe()
    Describe.to_csv("Describe.csv")
    
    print("Please check the folder")
```

```python
graph()
```

So, Improvization 2 will provide levarage to the user to enter folder name with every file name.
This will allow user to create different folder for every data file to store its statistics 

```python
def graph():
    
    import os
    import pandas as pd
    import numpy as np
    import matplotlib.pyplot as plt
    import seaborn as sns
    
    dir = input("Enter the path  : ")     # e.g : C:\\Users\\lenovo\\Desktop\\Praxis\\Practice_doc (without " ")
    os.chdir(dir)
    
    file = input('Enter the file name')   # e.g cars.csv (without " ")
    df = pd.read_csv(file) 
    
    # Allow user to enter folder name so that user can create multiple folder for different data  
    folder1 = input('Enter folder name ')
    os.mkdir(folder1)
    folder =  dir + '\\'+ '\\' + folder1
    os.chdir(folder)
    
    print("Print All if Column not specified ")
    col = input('Enter the name of column separated by comma : ')
    col1 = list(col.split(","))
    if col1 == ["All"]:
        col = []
        for i in df:
            col.append(i)
    else:
        col = col1
        
    for i in col:
        if (np.dtype(df[i]) != 'object' and len(np.unique(df[i]>50))):
            # Histogram
            df.hist(column=i,grid=False,figsize=(10,8))
            plt.savefig(str(i)+'_histogram.png')
            plt.show(str(i)+'_histogram')
            plt.close()

            # Boxplot
            df.boxplot(column=i,grid=False,notch=True,figsize=(10,8))
            plt.savefig(str(i)+'_boxplot.png')
            plt.show(str(i)+'_boxplot')
            plt.close()

        else:
            # Bar graph
            sns.countplot(df[i])
            plt.savefig(str(i)+'_bar_graph.png')
            plt.show(str(i)+'_bar_graph')
            plt.close()
            
    Describe = df.describe()
    Describe.to_csv("Describe.csv")
    
    print("Please check the folder")
```

```python
graph()
```
