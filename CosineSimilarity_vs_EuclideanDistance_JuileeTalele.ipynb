### Assumption :
- Different catogories differ from each other
- There is similarity within each article Ex. Maximum similarity between the Football players, but minimum similarity between football players and articles from Data Scince because they dont have any connection.

Lets find out either euclidean distance or cosine similarity gives better output

```python
# Importing require lib
import pandas as pd
import numpy as np
import os
import wikipedia
```

Function for Euclidean distance

```python
def euclidean_distance(x, y):   
    return np.sqrt(np.sum((x - y) ** 2))
```

Function for Cosine Similarity 

```python
def cosine_similarity(x, y):
    return np.dot(x, y) / (np.sqrt(np.dot(x, x)) * np.sqrt(np.dot(y, y)))
```
