### Central limit theoram

```python
def clt(n):
    mean_arr = []
    for j in range(0,10000):
        x = np.random.exponential(10,n)
        mean = np.mean(x)
        mean_arr.append(mean)
    print('Mean_array',np.mean(mean_arr))
    print('Std',np.std(mean_arr))
    x = np.random.exponential(10,n)
    sns.distplot(x)
    sns.distplot(mean_arr)
    
```

```python
clt(5)
```

```python
clt(100)
```

```python
clt(1000)
```

```python
clt(10000)
```

 ### Conclusion : As you increase the sample size the spread of se(mean_array) decreases
